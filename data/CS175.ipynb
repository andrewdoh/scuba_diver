# Data analysis for double Q-learning #

$$ \textbf{parameters: } \alpha = 0.1, \gamma =1, \epsilon=0.1 $$

```python
# data analysis and wrangling
import pandas as pd
import numpy as np
import random as rnd

# visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize']=(6,6)
```

```python
action_data = pd.read_csv('./3floors/actions.txt', sep=" ", header = None,names = ["episode", "action"])
time_data = pd.read_csv('./3floors/time.txt', sep=" ", header = None,names = ["episode", "time"])
death_data = pd.read_csv('./3floors/death.txt', sep=" ", header = None,names = ["episode", "alive"])
```

```python
sns.set_style("whitegrid")
ax = sns.barplot(x="action", y="episode", data=action_data).set_title('Distribution of Actions Taken(Map: Sideways_3floors)')
```

```python
# "Melt" the dataset to "long-form" or "tidy" representation
iris = pd.melt(action_data, "action", var_name="episode")
plt.rcParams['figure.figsize']=(6,6)
# Draw a categorical scatterplot to show each observation
sns.swarmplot(x="action", y="episode", data=action_data).set_title("Distribution of Actions")
```

```python
# Plot the orbital period with horizontal boxes
ax = sns.boxplot(x="episode", y="action", data=action_data,
                 whis=np.inf)
plt.rcParams['figure.figsize']=(6,6)
# Add in points to show each observation
ax = sns.stripplot(x="episode", y="action", data=action_data,
              jitter=True, size=3,linewidth=0).set_title('Boxplot of Actions')


```

```python
plt.rcParams['figure.figsize']=(6,6)
sns.set_style("whitegrid")
ax = sns.countplot(x="alive",data=death_data).set_title('Agent Deaths(4x4)')
```
