# Data analysis for double Q-learning #

$$ \textbf{parameters: } \alpha = 0.1, \gamma =1, \epsilon=0.1 $$

```python
# data analysis and wrangling
import pandas as pd
import numpy as np
import random as rnd

# visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize']=(6,6)
```

## 3floors ##

```python
action_data = pd.read_csv('./a=0.1,g=1,e=0.1/3floors/actions.txt', sep=" ", header = None,names = ["episode", "action"])
time_data = pd.read_csv('./a=0.1,g=1,e=0.1/3floors/time.txt', sep=" ", header = None,names = ["episode", "time"])
death_data = pd.read_csv('./a=0.1,g=1,e=0.1/3floors/death.txt', sep=" ", header = None,names = ["episode", "alive"])
reward_data = pd.read_csv('./a=0.1,g=1,e=0.1/3floors/rewards.txt', sep=" ", header = None,names = ["episode", "reward"])
rewards = reward_data['reward'].tolist()
```

### Cumulative reward / per episode ###

```python
sns.set_style("whitegrid")
plt.plot(range(1,len(rewards) + 1),rewards,'r-',label='reward/episode')
plt.axhline(y=95,xmin=0,xmax=50, color='g',linewidth=2.5,linestyle='--',label='Optimal')
plt.legend()
plt.xlabel('episodes')
plt.ylabel('cumulative reward')
plt.title('Double Q-learning(Map: Sideways 3 floors)')
plt.show()
```

### Distribution of Actions ###

```python
sns.set_style("whitegrid")
ax = sns.barplot(x="action", y="episode", data=action_data).set_title('Distribution of Actions Taken(Map: Sideways 3 floors)')
```

```python
# "Melt" the dataset to "long-form" or "tidy" representation
iris = pd.melt(action_data, "action", var_name="episode")
plt.rcParams['figure.figsize']=(6,6)
# Draw a categorical scatterplot to show each observation
sns.swarmplot(x="action", y="episode", data=action_data).set_title("Distribution of Actions(Map: Sideways 3 floors)")
```

```python
# Plot the orbital period with horizontal boxes
ax = sns.boxplot(x="episode", y="action", data=action_data,
                 whis=np.inf)
plt.rcParams['figure.figsize']=(6,6)
# Add in points to show each observation
ax = sns.stripplot(x="episode", y="action", data=action_data,
              jitter=True, size=3,linewidth=0).set_title('Boxplot of Actions(Map: Sideways 3 floors)')
```

### Death Count ###

```python
plt.rcParams['figure.figsize']=(6,6)
sns.set_style("whitegrid")
ax = sns.countplot(x="alive",data=death_data).set_title('Agent Deaths(Map: Sideways 3 floors)')
```
